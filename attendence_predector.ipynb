```python
import pandas as pd

# Let's create some dummy student data for Revnath Choure College
data = {
    'Student_ID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Classes_Attended': [40, 38, 15, 42, 20, 39, 10, 45],
    'Total_Classes': [45, 45, 45, 45, 45, 45, 45, 45],
    # 1 means the student Passed, 0 means they Failed
    'Passed': [1, 1, 0, 1, 0, 1, 0, 1] 
}

# Convert this dictionary into a pandas 'DataFrame' (which is just Python's version of an Excel table)
df = pd.DataFrame(data)

# Print out the table to make sure it looks right
df
```

```python
# Calculate the attendance percentage and add it as a new column
df['Attendance_Percentage'] = (df['Classes_Attended'] / df['Total_Classes']) * 100

# Show the updated table to make sure the math worked
print("--- Updated Data Table ---")
print(df)
print("\n")

# --- Preparing Data for the AI ---

# X is our Feature (what the AI uses to learn)
X = df[['Attendance_Percentage']] 

# y is our Target (the answer key: 1 = Pass, 0 = Fail)
y = df['Passed']

print("Data is successfully split into Features (X) and Target (y)!")
```

```python
from sklearn.linear_model import LogisticRegression
import pandas as pd

# 1. Create the AI model 
model = LogisticRegression()

# 2. Train the model. This is where the AI studies the data to find the pattern!
model.fit(X, y)
print("AI Model Training Complete! \n")

# 3. Let's test it! 
# What if a student has 72% attendance? Let's ask the AI if they will pass.
test_data = pd.DataFrame({'Attendance_Percentage': [72.0]})
prediction = model.predict(test_data)

# Print the result in plain English
if prediction[0] == 1:
    print("The AI predicts: This student with 72% attendance will PASS.")
else:
    print("The AI predicts: This student with 72% attendance will FAIL.")
```
